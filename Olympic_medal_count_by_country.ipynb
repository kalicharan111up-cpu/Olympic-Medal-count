# Insights

Here are some insights from the plot "Medals won by year":

1. **Overall Upward Trend**

The total number of medals awarded has generally increased over the years, showing the growing scale of the Olympics.

This is likely due to the addition of new sports/events and more participating countries.

2. **Early Years (18961940s)**

Fewer medals were awarded in the initial years, reflecting the smaller size of the Olympics back then.

The dips around 1916, 1940, and 1944 are due to the World Wars, when the Olympics were canceled.

3. **Post-War Boom (1950s1980s)**

After WWII, the number of medals increased steadily, showing the expansion of sports disciplines and greater global participation.

A noticeable rise in the 1960s1980s reflects the Olympics becoming a major international event.

4. **Modern Era (1990s2016)**

The medal count stabilizes at a higher level, typically above 2000 medals per Olympics.

This indicates the Olympics reached a mature, global scale with many disciplines standardized.

Small fluctuations occur due to the introduction or removal of sports.

5. **Key Observations**

Sharp drops (like in 1980 & 1984) may be linked to boycotts (Moscow 1980, Los Angeles 1984).

The overall rise reflects not just more athletes, but also a broadening of opportunities across sports.

# Top 5 dominating countries

```python
print("/nTop 5 dominating countries:")
print(df["NOC"].value_counts().head(6))
```

```python
# create a year wise analysis of top 5 countries
top_5_countries = df["NOC"].value_counts().head(5).index
top_5_df = df[df["NOC"].isin(top_5_countries)].pivot_table(
    index="Year", columns="NOC", values="Medal", aggfunc="count", fill_value=0
)
```

```python
# plot the bar chart for the countries that have won most medal won
plt.figure(figsize=(12, 6))
top_5_df.plot(kind="bar", stacked=True)
plt.xlabel("")
plt.ylabel("Medal")
plt.title("Top 5 countries that have most medal")
plt.legend(bbox_to_anchor=(1.05, 1), loc= 'upper left')
plt.tight_layout()
```

# Key observations:

**USA Dominance:**

The purple bars (USA) are consistently the largest across most years, showing that the USA has been the most dominant medal winner historically.

Peaks are especially visible around the 1980s2000s.

**European Countries:**

Germany (GER, green) and Great Britain (GBR, orange) show strong performances, particularly from the mid-20th century onwards.

Germany shows fluctuations (partly due to political events like East/West Germany before reunification).

**France (FRA, blue) and Italy (ITA, red):**

These two countries contribute steadily but generally remain behind the USA and Germany.

France shows a consistent presence since the beginning (as the Olympics started in Athens 1896 and Paris hosted early on).

Italy had notable spikes around the 1930s and early 2000s.

```python
# participation trend
plt.figure(figsize=(12, 6))
df.groupby("Year")["Name"].count().plot(kind="line")
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Participation trend")
plt.show()
```

# Key observations:

**Overall Growth:**

There is a clear upward trend in participation from the early Olympics until recent years.

Early years had fewer than 2,000 participants, but participation steadily grew to over 15,000 athletes in modern Olympics.

**Key Growth Phases:**

Sharp increases are visible around the 1950s1980s, likely due to:

More countries joining the Olympics after WWII.

Inclusion of new sports and womens events.

A major rise is also seen in the 1990s, possibly due to globalization and professional athletes competing.

**Dips in Participation:**

Noticeable drops during the World War periods (1916, 1940, 1944 Olympics canceled) and around 1980 & 1984, which were affected by political boycotts.

The zig-zag fluctuations after the 1990s suggest that the dataset may include both Summer and Winter Olympics (hence alternating lower and higher participation counts).

**Recent Era:**

Participation remains high, stabilizing around 10,00012,000 athletes per Summer Olympics.

Winter Olympics generally have fewer participants, hence the sharp falls in alternating years.

# Key observations:

**Top Sports by Participation:**

Athletics (Track & Field) is by far the most participated sport, with the highest count exceeding 35,000.

Followed by Gymnastics 27000 and Swimming 23000.

These three dominate because they include many disciplines and attract athletes from nearly every participating country.

**Strong Participation Sports:**

Other widely represented sports are Rowing, Fencing, Cycling, Cross-Country Skiing, Alpine Skiing, and Wrestling, all with counts above 10,000.

This shows strong representation in both Summer (rowing, fencing, cycling) and Winter (skiing, ice skating) games.

**Mid-Level Sports:**

Sports like Boxing, Ice Hockey, Basketball, Handball, and Judo have moderate participation, ranging around 4,0007,000.

**Low Participation Sports:**

Niche or recently added sports like Trampoline, Rugby Sevens, Beach Volleyball, and Taekwondo show lower counts.

Obsolete or demonstration sports (e.g., Tug of War, Polo, Lacrosse, Military Ski Patrol, Aeronautics) appear with very low counts, reflecting their limited or discontinued presence in the Olympics.
